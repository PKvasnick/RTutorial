```
library(tidyverse)
library(scales)
set.seed(123)
n = 1000

Education = rbinom(n, 2, 0.5)
Neuroticism = rnorm(n) + Education
Salary = Education * 2 + rnorm(n) - Neuroticism * 0.3

Salary = sample(10000:11000,1) + rescale(Salary, to = c(0, 100000))
# summary(Salary)
Neuroticism = rescale(Neuroticism, to = c(0, 7))
# summary(Neuroticism)
Education = factor(Education, labels = c("Low", "Medium", "High"))

data <- data.frame(
  Salary,
  Neuroticism,
  Education
)
ggplot(data, mapping=aes(x=Neuroticism, y=Salary)) + geom_point() + geom_smooth(method = "lm")
```

```
ggplot(data, mapping=aes(x=Salary, y=Neuroticism, color = Education)) + geom_point() + geom_smooth(method = "lm")
```

```
write.csv(data,"Neuro.csv", row.names = FALSE)
```
